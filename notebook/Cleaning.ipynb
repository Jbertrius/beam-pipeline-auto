```python
import pandas as pd
```

```python
_all = pd.read_csv('dataTrain.csv')
```

### Suppression des doublons

```python
_all = _all.drop_duplicates()
display(_all.shape)
```

```python
_all = _all.sort_values(by=['vin','idvariante'])
display(_all.shape)
```

### Suppression de VIN contenant I,O ou Q

```python
_all = _all[ ~_all['vin'].str.contains('[IOQ]') ]
display(_all.shape)
```

### Suppression de VIN ayant plus d'une variante

```python
incorrect_vin = _all.groupby('vin').filter(lambda x: len(x)>=2)
```

```python
nb_vin_redundant = len(incorrect_vin['vin'].unique())
```

```python
vin_redundant = incorrect_vin['vin'].unique().tolist()
```

```python
vin_redundant
```

```python
_all = _all[ ~_all['vin'].isin(vin_redundant) ]
display(_all.shape)
```

### Suppression de VIN ayant un mauvais format de Serial number

```python
_all = _all[ _all['vin'].str.contains('\w{11}[0-9]{6}') ]
```

```python
display(_all.shape)
```

```python
_all.to_csv( 'dataTrain_corrected.csv', index=False)
```
